# Xử lý dữ liệu

```python
# Sao chép cột 'weather' vào cột mới 'current_weather'
data['current_weather'] = data['weather'].copy()

# Đổi vị trí cột 'current_weather' và cột 'weather'
data = data.reindex(columns=['month', 'date', 'time', 'temperature', 'rain(mm)', 'rain_percent', 'cloud_percent', 'pressure(mb)', 'wind(km/h)', 'gust(km/h)', 'current_weather', 'weather'])
```

```python
# Biến đổi giá trị của cột cuối cùng thành giá trị của dòng trước đó
data.iloc[:, -1] = data.iloc[:, -1].shift(-1)
data
```

# Loại bỏ thuộc tính dư thừa
Ở đây chúng ta thấy cột time và date có thể loại bỏ vì giá trị của nó không có nhiều ảnh hưởng đến dự đoán thời tiết.

```python
# data.drop('month', axis=1, inplace=True)
data.drop('date', axis=1, inplace=True)
data.drop('time', axis=1, inplace=True)
data.drop('rain_percent', axis=1, inplace=True)
data
```

Xoá hàng có giá trị None

```python
data = data.dropna()
data
```

# Xử lý dữ liệu nhãn

khảo sát giá trị

```python
label_value = data["weather"].unique()
label_value
```

biến đổi

```python
# Tạo bản đồ ánh xạ giữa giá trị ban đầu và giá trị số tương ứng
label_mapping = {value: i for i, value in enumerate(label_value)}

# Áp dụng ánh xạ để chuyển đổi cột "weather" thành giá trị số
data["weather"] = data["weather"].map(label_mapping)
data["current_weather"] = data["current_weather"].map(label_mapping)

data
```
