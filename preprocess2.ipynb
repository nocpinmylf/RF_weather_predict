## Các giá trị thời gian

```python
def convert_time_val(time_):
  if time_.split(' ')[0].strip() == 'No':
    return 0
  hour = datetime.strptime(time_, '%I:%M %p').time().hour
  mins = datetime.strptime(time_, '%I:%M %p').time().minute
  hour = hour + 1 if mins > 50 else hour
  return hour
```

### khảo sát giá trị sunrise

```python
sunrise_value = data["sunrise"].unique()
sunrise_value
```

```python
data['sunrise'] = data['sunrise'].apply(lambda x: convert_time_val(x))
```

### khảo sát giá trị sunset

```python
sunset_value = data["sunset"].unique()
sunset_value
```

```python
data['sunset'] = data['sunset'].apply(lambda x: convert_time_val(x))
```

### khảo sát giá trị moonrise

```python
moonrise_value = data["moonrise"].unique()
moonrise_value
```

```python
data['moonrise'] = data['moonrise'].apply(lambda x: convert_time_val(x))
```

### khảo sát giá trị moonset

```python
moonset_value = data["moonset"].unique()
moonrise_value
```

```python
data['moonset'] = data['moonset'].apply(lambda x: convert_time_val(x))
```
